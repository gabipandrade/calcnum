Na primeira questão, o resultado foi o vetor c o qual foi feito a partir do somatório de dois vetores randômicos a e b de n dimensões com dois escalares alfa e beta também aleatórios, isso feito para o iem (a).Já para o item (b) é feito uma matriz randômica A de ordem n tal que se calcula a m-émissa potência. Para o seguinte item, é feita a análise para os casos m igual a 2, 3 e 4, respectivamente. Conforme tanto para o caso do aumento da dimensão de n vetorial quanto para o aumento escalar da multiplicação matricial, nota-se que aumento crescente do tempo, principalmente para os casos de maior grau de n tal qual ocorre com a ordem 4 uma vez que este é o que apresenta maior número de operações a serem realizadas.

```python
import time
import numpy as np
import matplotlib.pyplot as plt
#Função recebe A e B e calcula o tempo para multiplicar c=alpha*A+ beta*B
def mult(A,B):
    alpha = np.random.rand(1)
    beta = np.random.rand(1)
    
    ti = time.time()
    
    c=alpha*A + beta*B
    
    return time.time() - ti

#n=1000000-10000000
#cria A e B aleatorios num for de 1000000 ate 10000000 de 100000 em 100000 e armazena os tempos de demora e o tamanho
tempos1,n1=[],[]
for i in range(1000000,10000000,100000):
    A=np.random.rand(i)
    B=np.random.rand(i)
    tempos1.append(mult(A,B))
    n1.append(i)
#cria o grafico do tamanho em funçao do tempo
fig, ax = plt.subplots(1,2,figsize=(15,4))
print(tempos1)
ax[0].plot(n1, tempos1, "-.r")
ax[0].set(xlabel="n", ylabel="Tempo [sec]", 
       title="n vs tempo")
ax[0].grid()
#cria o grafico loglog do tamanho em função do tempo 
ax[1].plot(n1, tempos1, "-.r")
ax[1].set(xlabel="n", ylabel="Tempo [sec]", 
       title="n vs tempo Loglog")
ax[1].grid()
ax[1].loglog()
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt
#recebe n:tamanho da matriz e m:potencia da multiplicação e retorna o tempo para que seja feita a multiplicação
def multmat(n,m):
    A = np.random.rand(n,n)
    C=A
    ti = time.time()
    for i in range(1,m,1):
        C = C @ A
    
    return time.time() - ti
#n:tamanho da matriz
#m:potencia de multiplicação


#m=2,3,4
#n variavel


#n=100-2000, m=2
tempos1,n1=[],[]
for i in range(100,2000,80):
    tempos1.append(multmat(i,2))
    n1.append(i)

#n=100-2000, m=3
tempos2,n2=[],[]
for i in range(100,2000,80):
    tempos2.append(multmat(i,3))
    n2.append(i)
    
#n=100-2000, m=4
tempos3,n3=[],[]
for i in range(100,2000,80):
    tempos3.append(multmat(i,4))
    n3.append(i)

fig, ax = plt.subplots(1,2,figsize=(15,4))
#grafico dos tamanhos n das matrizes potencias de 2 em vermelho, potencias de 3 em azul e potencias de 4 em preto em função do tempo
ax[0].plot(n1, tempos1, "-.r", n2,tempos2, "-.b", n3,tempos3,"-.k")
ax[0].set(xlabel="n", ylabel="Tempo [sec]", 
       title="n vs tempo")
ax[0].grid()
#grafico log log dos tamanhos n das matrizes de potenciais de 2 em vermelho, potencias de 3 em azul e potencias de 4 em preto em função do tempo
ax[1].plot(n1, tempos1, "-.r", n2,tempos2, "-.b", n3,tempos3,"-.k")
ax[1].set(xlabel="n", ylabel="Tempo [sec]", 
       title="n vs tempo Loglog")
ax[1].grid()
ax[1].loglog()
plt.show()
```

Nota-se que os gráficos bidimensionais apresentam projeção coerente ao tridimensional original uma vez que vetores gradientes compõe s inclinação da curva e a mesma, por conseguinte, foram o nível de contorno das zonas extremo inferior e superior ao eixo z em seu ponto originário.

```python
from mpl_toolkits import mplot3d
import matplotlib.cm as cm
import numpy as np
import matplotlib.pyplot as plt
#função recebe x e y e retorna a multiplicação 3cos(x)cos(y) 
def f(x,y):
  return 3*np.cos(x)*np.cos(y)
#função recebe x e y e calcula a derivada direcional em x da função 3cos(x)cos(y)
def derv_x(x,y):
  return -3*np.sin(x)*np.cos(y)
#função recebe x e y e calcula a derivada direcional em y da função 3cos(x)cos(y)
def derv_y(x,y):
  return -3*np.cos(x)*np.sin(y)

x=np.linspace(0, 2*np.pi, num=50)
y=np.linspace(0, 2*np.pi, num=50)
x,y=np.meshgrid(x,y)
z=f(x,y)

#grafico da 3d da superficie da função 3cos(x)cos(y) em termos de x e y
fig=plt.figure(figsize=(10,5))
ax= plt.axes(projection="3d")
ax.plot_surface(x, y, z, cmap='hsv', edgecolor=None)
ax.set(xlabel='$x$', ylabel='$y$', zlabel='$z$')
m = cm.ScalarMappable(cmap=cm.jet)
m.set_array(z)
plt.colorbar(m)
ax.set_title('Grafico 3d da função')
plt.show()
#grafico de curva de nivel
fig=plt.figure(figsize=(10,5))
ax=plt.contour(x,y,z,cmap='jet')
plt.show()
#curva de gradiente feito pelas derivadas direcionais
fig=plt.figure(figsize=(10,5))
a=derv_x(x,y)
b=derv_y(x,y)
plt.quiver(x,y,a,b)
plt.show()
#graficos de curva de nivel e curva de gradiente juntos
fig=plt.figure(figsize=(10,5))
plt.quiver(x,y,a,b)
ax=plt.contour(x,y,z,cmap='jet')
plt.show()
```
